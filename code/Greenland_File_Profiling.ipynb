```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
folder_root = '/content/drive/MyDrive/Semester3/Capstone/Greenland2023_data_drone'

#get list of files
files = glob.glob(folder_root+'/**/*', recursive=True)
print(files[0:5])
#for f in files[0:3]:
  #print((f))
  #os.path.splitext(files[0])
```

```python
#add details about the files in a dataframe
df = pd.DataFrame(files, columns=['full_path'])
df['ext'] = df['full_path'].apply(lambda x: os.path.splitext(x)[1].lower())
df['capstone'] = df.full_path.str.contains("Capstone_") #rough way to see if related to the 3 folders for our project
df['spectra'] = np.where(df.full_path.str.contains("_D."), 'RGB',
                np.where(df.full_path.str.contains("_NIR."), 'Near IR',
                np.where(df.full_path.str.contains("_RE."), 'Red Edge',
                np.where(df.full_path.str.contains("_R."), 'Red',
                np.where(df.full_path.str.contains("_F."), 'NDVI',
                np.where(df.full_path.str.contains("_G."), 'Green', 'Other'))))))
df['height'] = np.where(df.full_path.str.contains("_top_height"), 'Top',
               np.where(df.full_path.str.contains("_medium_height"), 'Medium',
               np.where(df.full_path.str.contains("_low_height"), 'Low', '')))
df['processed'] = df.full_path.str.contains("/Processed/") #rough estimated way to determine
df['mosaic'] = df.full_path.str.contains("/2_mosaic/") #rough estimated way
df['tiles'] = df.full_path.str.contains("/tiles/") #rough estimated way
df
```

```python
df[df.ext != ''].groupby(['ext']).capstone.count().sort_values(ascending=False)
```

```python
#summarize some things
df[df.ext != ''].groupby(['capstone']).capstone.count()  #7200 files for capstone (16000 total)

df[(df.ext != '') & (df.ext.isin(['.png','.dng','.tif','.jpg']))].groupby(['capstone']).capstone.count() #6696 images

df_filter = df[(df.ext != '') & (df.ext.isin(['.png','.dng','.tif','.jpg'])) & (df.capstone == 1)] #for just our images
display(df_filter.groupby(['ext']).full_path.count()) #mostly jpg and tif files
print('\n')
display(df_filter.groupby(['spectra']).full_path.count())
```
