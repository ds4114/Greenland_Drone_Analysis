### Image data ingestion/prep

```python
def make_image_arr():
    img_root_folder = r'C:\Users\Devan\Downloads\greenland' 
    
    altitude_prefix = [('low','map3a',)] #low only now
    channel = ["green", "red", "red edge", "nir"]
    ims = []
    for p in altitude_prefix:
        for c in channel:
            filepath = os.path.join(img_root_folder, p[1] + "_transparent_mosaic_" + c + "_" + p[0] + ".tif") #ex "map3a_transparent_mosaic_green_low.tif"
            ims.append(cv2.imread(filepath, cv2.IMREAD_GRAYSCALE).astype(np.float32))
        
    return ims

low_raw = make_image_arr()
```

```python
def clean_image(img_arr):
    #attempt to make this set easier to work with using xarray instead. 
    #Doing this manually but will need to update when we add more altitudes and fix issues with different sizes
    
    dim1 = img_arr[0].shape[0]
    dim2 = img_arr[0].shape[1]
    yrange = range(0,-1*dim1, -1) #not sure what to agree on for coords but putting (0,0) as top left and making y negative to stay cartesian
    xrange = range(0,dim2)
    temp1 = xr.DataArray(img_arr[0], coords={'altitude':'low','channel':'green'   , 'y': yrange,'x': xrange}, dims=["y", "x"], name='lgxy')
    temp2 = xr.DataArray(img_arr[1], coords={'altitude':'low','channel':'red'     , 'y': yrange,'x': xrange}, dims=["y", "x"], name='lrxy')
    temp3 = xr.DataArray(img_arr[2], coords={'altitude':'low','channel':'red_edge', 'y': yrange,'x': xrange}, dims=["y", "x"], name='lrexy')
    temp4 = xr.DataArray(img_arr[3], coords={'altitude':'low','channel':'nir'     , 'y': yrange,'x': xrange}, dims=["y", "x"], name='lnxy')
    low_xr = xr.concat([temp1, temp2, temp3, temp4], dim='channel').rename('low_4')
    
    low_xr = low_xr.where( low_xr.any(dim = ['channel']) ) #make NaN when all zero across a channel (and altitude) to indicate black space
    #print(low_xr)
    
    #other cleaning? - ex. divide by 255?; add a random value for subset sampling later?
    
    return low_xr

#low_xr = clean_image(low_raw)
```

```python
def get_spec_profile(arr, x=0,y=0):
    return [arr[0][x][y], arr[1][x][y], arr[2][x][y], arr[3][x][y]] #green, red, RE, NIR 


drone_wavelengths = [560, 650, 730, 860] #https://ag.dji.com/mavic-3-m/specs
test1 = get_spec_profile(low_raw, 2000, 2000)
test2 = get_spec_profile(low_raw, 5500, 5000)
test3 = get_spec_profile(low_raw, 6000, 5000)
test4 = get_spec_profile(low_raw, 8000, 6000)
sns.lineplot(x=drone_wavelengths, y=test1);
sns.lineplot(x=drone_wavelengths, y=test2);
sns.lineplot(x=drone_wavelengths, y=test3);
sns.lineplot(x=drone_wavelengths, y=test4);
```
